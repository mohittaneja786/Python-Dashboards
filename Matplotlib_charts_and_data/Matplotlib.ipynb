```python
import seaborn as sns

plt.figure(figsize = (10,6))
sns.histplot (
                df['salary'],
                bins = 20,
                kde = True,
                color = 'Teal',
                edgecolor = 'black'   
) 

plt.title('Salary Distribution Accross Employee', fontsize = 14)
plt.xlabel('Salary', fontsize = 12)
plt.ylabel('Count', fontsize = 12)
plt.grid(True, alpha = 0.3)
plt.savefig("1.png")
plt.show()

```

## 2. Salary comparison based on promotion status

```python
df
```

```python


employee_melted_data = pd.melt(
            df,
            id_vars = ["promotion_hr_based"],    
            value_vars = ['salary','final_incremented_salary'],
            var_name = 'Salary_Type',
            value_name = 'Salary'    
)

plt.figure(figsize = (8,4.5))
sns.barplot(
            x = 'promotion_hr_based',
            y = 'Salary',
            hue = 'Salary_Type',
            palette = 'Set2',
            data = employee_melted_data)

plt.title('Salary Promotion By Promotion Status', fontsize = 14)
plt.xlabel('Promotion HR data', fontsize = 12)
plt.ylabel('Salary', fontsize = 12)
plt.legend(title = 'Salary Type')
plt.savefig("2.jpg")
plt.show() 

```

# 3. Increment percentage evluation

```python
plt.figure (figsize = (10,6))

sns.boxplot(
            x = 'promotion_hr_based',
            y = 'increment_percentage',
            hue = 'promotion_hr_based',
            palette = 'Pastel1',
            data = df
)


plt.title('Increment Percentage', fontsize=14)
plt.xlabel ('Promotion HR based', fontsize = 12)
plt.ylabel ('Increment Percentage (%)', fontsize = 12)
plt.legend (title = 'Promotion Plan')
plt.savefig("3.jpg") 
plt.show()   
   
           
```

```python
df.columns 
```

# 4. Relationship between expereince and salary growth 

```python
plt.figure (figsize = (10,6))

sns.scatterplot (
                x = 'experience_years',
                y = 'salary',
                hue = 'promotion_hr_based',
                size = 'increment_percentage',
                style = 'promotion_hr_based',
                palette = 'deep',
                data = df,
                sizes = (50,200)  
)

plt.title('Experience VS Year By promotion cycle', fontsize=14)
plt.xlabel ('Year of Experience', fontsize = 12)
plt.ylabel ('salary', fontsize = 12)
plt.legend (title = 'Promotion Status')
plt.savefig("4.png", bbox_inches = 'tight')    
plt.show()           
       
```

```python
df.columns
```

# 5. Sales VS Productivity

```python
plt.figure (figsize = (10,6))

sns.scatterplot (
                x = 'time_productive',
                y = 'sales',
                hue = 'experience_years',
                size = 'experience_years',
                #style = 'promotion_hr_based',
                palette = 'viridis',
                data = df,
                sizes = (50,200)  
)

plt.title('Sales V/s Produtivity', fontsize=14)
plt.xlabel ('Productivity Time', fontsize = 12)
plt.ylabel ('Sales', fontsize = 12)
plt.legend (title = 'Promotion Status')
plt.savefig("5.png", bbox_inches = 'tight')    
plt.show()           
       
```

```python
df.columns
```

## 6. Average sales and productivity by experience

```python
grouped_data = df.groupby('experience_years')[['sales','time_productive']].mean().reset_index()

print("Grouped data (Average Sales and Productivity : \n", grouped_data)
```

## 7. Top performers

```python
top_performers = df [['first_name', 'sales']].sort_values(by = 'sales', ascending = False).head(10)
plt.figure (figsize = (10,6))

sns.barplot (
                x = 'first_name',
                y = 'sales',
                data = top_performers,
                palette = 'Blues',
)

plt.title('Top 10 Performers', fontsize=14)
plt.xlabel ('Employee Name', fontsize = 12)
plt.ylabel ('Sales', fontsize = 12)
plt.xticks (rotation = 15)
plt.savefig("7.png", bbox_inches = 'tight')    
plt.show()     
```

```python
top_performers
```

## 8. Experience distribution

```python
plt.figure(figsize = (8,5))

sns.histplot( df['experience_years'],
                bins = 20,
              color = 'Teal', 
              kde = True
)

plt.title('Distribution in Employee Experience', fontsize=14)
plt.xlabel ('Experience Years', fontsize = 12)
plt.ylabel ('Counts', fontsize = 12)
plt.xticks (rotation = 15)
plt.savefig("8.png", bbox_inches = 'tight')    
plt.show() 
```

## 9. Promotion hours vs experience

```python
promotion_data = df.groupby('experience_years')[['increment_percentage']].mean().reset_index()

print("Promotion data : ", promotion_data)
```

```python
promotion_data = df.groupby('experience_years')[['increment_percentage']].mean().reset_index()

plt.figure(figsize = (8,6))
sns.lineplot(data = promotion_data,
                x = 'experience_years',
                y = 'increment_percentage',
                palette = 'Set2',
                marker = 'o'
)


plt.title('Promotion hours vs experience', fontsize=14)
plt.xlabel ('Experience Years', fontsize = 12)
plt.ylabel ('Promotion : Productive Hours', fontsize = 12)
plt.savefig("9.png", bbox_inches = 'tight')    
plt.show() 
```

## 10. Bubble Chart (Experience VS sales)
